#### This gives us the unemployment data for the state of california.  Now, let's use python to obtain this data and download it.  First we assign, a string to the relevent url.

```python
bls_url ='https://download.bls.gov/pub/time.series/la/la.data.21.indiana'
```

#### Now we use 'urllib.request' and 'urlopen' functions to get the source code for the Indiana unemployment data.

```python
Source_Code = urllib.request.urlopen(bls_url).read().decode("utf-8")
```

#### Let's look at the output from this read.  It appears that \t seperates columns and \n seperates rows.

```python
Source_Code[0:1000]
```

#### Now that we have the Source Code, which in this case is a formatted table, we can export it to a txt file on our directory. We use the 'open' function to open a file in our directory.  The open command here takes two inputs, the first is the file name, and the second is 'w', which means we want to write to this file.  We can also use 'a', to append the file, and 'r' to read from the file.

```python
output_file = open("Files_Directory/BLS/bls_1.txt","w")
```

#### Now we have an object called output_file, and we want to write our source code to the output file object, which will, in turn, write it to the file on our directory.

```python
output_file.write(Source_Code)
output_file.close()
```

#### If you check your directory, you should have a file called bls_1.txt which will have all of the unemployment statistics for the state of Indiana.
